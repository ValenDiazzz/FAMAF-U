```python
import numpy as np
from ipynb.fs.full.Ej12 import cholesky
from ipynb.fs.full.Ej3 import sol_trinffil
from ipynb.fs.full.Ej3 import sol_trsupfil
```

```python
A1 = np.array([
    [4, 0],
    [0, 9]
],    dtype="float") 
b1 = np.array([8, 18],  dtype="float")
```

```python
def sol_defpos(B,y):
    A=B.copy()
    b=y.copy()
    G=cholesky(A)
    y = sol_trinffil(G.T,b)
    x=sol_trsupfil(G, y)
    return x
```

```python
sol_defpos(A1,b1)
```
