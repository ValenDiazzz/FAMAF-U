```python
import numpy as np
```

```python
Csup = np.array([
    [4, 2, 6],
    [2, 2, 5],
    [6, 5, 29],
], dtype="float")
b3 = np.array([18, -2, 7], dtype="float")
```

```python
#outer product algorithm
def cholesky(M):
    A=M.copy()
    assert np.array_equal(A, A.T), "La matriz no es simetrica"
    n = A.shape[0]
    G=np.zeros((n,n))
    
    for i in range(n):
        assert A[i, i]>0, "La matriz no es definida positiva"
        G[i, i] = np.sqrt(A[i, i])
        #J =i+1:n
        G[i, i+1:n] = A[i, i+1:n]/G[i, i]
        A[i+1:n, i+1:n] = A[i+1:n, i+1:n] - np.outer(G[i, i+1:n], G[i, i+1:n]) #Al primero lo agarra y lo toma como una fila y al segundo como una columna. Mejor que usar reshape
        
    return G
```

```python
cholesky(Csup)
```

```python
#inner product algorithm
def cholesky_ip(M):
    A=M.copy()
    assert np.array_equal(A, A.T), "La matriz no es simetrica"
    n=A.shape[0]
    for i in range(n):
        for k in range(i):
            A[i,i]=A[i,i]-(A[k,i])**2
        
        assert  A[i, i]>0,"La matriz no es definida positiva"
        A[i,i]=np.sqrt(A[i,i])
        
        for j in range(i+1,n):
            for k in range(i):
                A[i,j]=A[i,j]-(A[k,i]*A[k,j])
            A[i,j]=A[i,j]/A[i,i]
    return np.triu(A)
```

```python
cholesky_ip(Csup)
```
