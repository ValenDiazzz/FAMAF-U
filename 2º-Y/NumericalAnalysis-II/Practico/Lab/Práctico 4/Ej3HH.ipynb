```python
#Defino al reflexion de householder que hace la reflexion a un vector
def hh(x):
    u= x.copy()
    n= len(u)
    b= max(np.abs(u))
    if b==0:
        g= 0
        t=1 #Nada m√°s la declaro para q no me genere problemas el return
    else:
        u= u*(1/b)
        t= np.linalg.norm(u,2)

        if u[0]<0:
            t=-t
        u[0]= u[0]+t
        g= u[0]/t
        u[1:n]= u[1:n]/u[0]
        u[0]= 1
        t= b*t
    return (u,g) #puede devolver t
```

```python
def qrhholder(A):
    m, n = A.shape
    Q = np.eye(m)
    R = A.copy()
    p = min(m, n)

    for j in range(p):
        # I = j:, J = j:
        u, rho = hh(R[j:, j])
        w = rho * u
        R[j:, j:] = R[j:, j:] - np.outer(w, u.T @ R[j:, j:])
        Q[:, j:] = Q[:, j:] - Q[:, j:] @ np.outer(w, u)

    return Q, R
```
