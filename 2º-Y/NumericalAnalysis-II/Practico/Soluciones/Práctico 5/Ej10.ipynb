```python
import numpy as np
```

```python
def sol_gradcon(M,y, it):
    b=y.copy()
    A=M.copy()
    n=A.shape[1]
    x=np.ones(n)
    r0=b-(A@x)
    p=r0.copy()
    k=0
    while np.linalg.norm(r0,2)>=1e-8 and k<it:
        v=np.dot(A, p)
        a=np.inner(r0,r0)/ np.inner(v,p)
        x=x+(a*p)
        r1=r0-(a*v)
        beta=np.inner(r1,r1)/np.inner(r0,r0)
        p=r1+(beta*p)
        r0=r1.copy()
        k+=1
    return x
```
