**Ejercicio 1**

Definir la función

        triangulo_iso(ang: float, largo: float)

que dibuja un triángulo isosceles con _ángulo en el vértice_ igual  a `ang` y con los lados iguales de longitud `largo`. El parámetro `ang` debe ingresarse en grados sexagesimales.    

_Observación._ El ángulo formado por los lados de igual longitud se llama `ángulo en el vértice` y el lado opuesto a él, `base`.

```
# Escriba el código más abajo

#Dibujar un triangulo isoceles sabiendo el tamaño de los lados iguales y su angulo

def triangulo_iso (angle, side):
  pass
  
  #medida de cada angulo adyacente a la base
  angulos = (180 - (180 - angle) / 2)

  #angulo en radianes
  gama = ((angle/180)*pi)

  #Tamaño de la base calculado con teorema del coseno
  base = sqrt((2*side*side) - (2*side*side*cos(gama)))

  #ejecucion del dibujo
  forward(side)
  right(angulos) 
  forward(base)
  right(angulos)   
  forward(side)
  right(180)
  
```

**Ejercicio 2**

Definir la función 

        poligono_regular(n: int, radio: float)
        
que dado un entero positivo `n`, y un número real `radio` dibuje un polígono regular de `n` lados de radio `radio` y todos los radios. 

_Nota:_ El _radio_ de un polígono regular es la longitud del segmento que une el centro del polígono con uno de sus vértices. Un _radio_ es un segmento que une el centro del polígono con uno de sus vértices. En  el ejercicio se pide dibujar un polígono regular con todos sus radios. 

_Ayuda:_ Si hiciste bien el ejercicio 1, este debería poder escribirse en pocas líneas, dado que cada lado del polígono con sus dos radios dibujados determina un triángulo isóceles. 


_Ejemplo_

- Si ejecuto `poligono_regular(3, 100)` obtengo

pol_3_100.png

- Si ejecuto `poligono_regular(7, 100)` obtengo

pol_7_100.png

- Si ejecuto `poligono_regular(6, 100)` obtengo

pol_6_100.png

```
# Escribir el código más abajo (en esta celda)

#Dibuja un poligono de lados n y cierto radio

def poligono_regular(n: int, ratio: float):

  #formula para saber caunto mide cada uno de los angulos interiores de un poligono regular
  angulo_interior = (((n-2)*180) / n)

  #cálculo de angulo exterior
  angle = (180 - angulo_interior)

  #ejecucion n veces de un triangulo isóceles
  for _ in range (n):
    triangulo_iso(angle, ratio)

    
```

**Ejercicio 3**

Definir la función 

        escalera_regular(n: int, radio: float)
        
que dado un entero positivo `n`, y un número real `radio` dibuje una "escalera caracol" de `n` escalones  vista desde arriba. La escalera está formada por triángulos isósceles semejantes con altura creciente y proporcional al radio (ver los ejemplos más abajo). 


_Ayuda:_ Basándote en el ejercicio 2, a la función `escalera_regular()` la podrías hacer copiando y modificando la definición de `poligono_regular()`  


_Ejemplos_

- Si ejecuto `escalera_regular(9, 25)` obtengo una escalera caracol de 9  escalones donde el primer escalón "mide" (de la base al vértice, por sus lados de igual longitud) `25`  y el último mide `225`.

esc_9_25.png

- Si ejecuto `escalera_regular(5, 30)` obtengo una escalera caracol de 5  escalones donde el primer escalón mide `30`  y el último mide `150`.

esc_5_30.png

- Si ejecuto `escalera_regular(3, 50)` obtengo una escalera caracol de 3  escalones donde el primer escalón mide `50`  y el último mide `150`.

esc_3_50.png

```
# Escribir el código más abajo (en esta celda)
def escalera_regular(n: int, ratio: float):
  
  #Tasa de crecimiento constante
  h = ratio
  
  #formula para saber caunto mide cada uno de los angulos interiores de un poligono regular
  angulo_interior = (((n-2)*180) / n)

  #cálculo de angulo exterior
  angle = (180 - angulo_interior)

  #ejecucion n veces de un triangulo isóceles
  for _ in range (n):
    triangulo_iso(angle, ratio)
    ratio = ratio + h
```
