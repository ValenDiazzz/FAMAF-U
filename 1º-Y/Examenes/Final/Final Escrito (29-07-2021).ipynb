**Ejercicio 1.** Escribir una función `numero_interior()` que dados tres números  enteros `p`, `q` y `n` devuelva `True` (el valor booleano) si `p` es menor o igual a `q` y `n` está entre  `p` y `q` (y puede ser `p` o `q`). En caso contrario, devuelve `False`.   

````python
# Escribir el código en esta celda

def numero_interior(p,q,n):
  #pre: recibe tres números enteros
  #post: devuelve True si p <= q y p<n<q o False en caso contrario
  assert type (p) == type (q) == type (n) == int , "Error, p, q y n deben ser números enteros"
  #escribo las condiciones que deben cumplir para ser verdaderas
  condicion_1 = p <= q
  condicion_2 = p <= n <= q
  
  return condicion_1 and condicion_2
````

````python
# Tests: la función debe pasar los siguientes tests

test(1, numero_interior(35, 43, 100), 'False')
test(2, numero_interior(35, 63, 51), 'True')
test(3, numero_interior(63, 35, 5), 'False')
test(4, numero_interior(-35, -43, -38), 'False')
test(5, numero_interior(-6, -4, -4), 'True')
````

**Ejercicio 2.** Escribir la función `minimo_positivo()` que encuentra el menor entero positivo de una lista de números enteros. Incluir en la función un `assert` que compruebe que es una lista de enteros. La función debe iterar sobre la lista de enteros y debe ser implementada teniendo en cuenta que solo se permite, a nivel de operaciones, métodos o funciones en listas, acceder a los elementos de una lista por el índice.   

No está permitido utilizar `sort()`, `sorted()` o métodos similares. Tampoco se puede usar la función `min()`, tanto en lístas como en pares de números (o una función similar o análoga).

````python
def minimo_positivo(lista):
  #pre: recibe una lista de numeros enteros
  #post: devuelve el menor entero postivo
  assert type (lista) == list and all(type (x) == int for x in lista)

  lista_p = []

  for x in lista:
    if 0 < x:
      lista_p.append(x)
    

  if lista_p != []:
    min = lista_p[0]
    for i in range(1,len(lista_p)):
      if lista_p[i]<min:
        min = lista_p[i]
  else:
    min = None

  return min
````

````python

# Tests: la función debe pasar los siguientes tests

test(1, minimo_positivo([10, 4, 1, 4, -10, -50, 32, 21]), '1')
test(2, minimo_positivo([10, 0, 4, 2, 0]), '2')
test(3, minimo_positivo([-10, 0, -4, -2, 0]), 'None')
````

**Ejercicio 3.** Escribir una función `extraer_coordenadas()` que dada una lista de números enteros, devuelve otra lista que es el resultado de extraer las coordenadas que son múltiplos de  `3` y `7` de la lista original. Por ejemplo

```
extraer_coordenadas([8,5,3,11,0,1,21,22,23,31,5,6,6,3,4,0,3,2])
```
devuelve la lista `[8, 11, 21, 22, 31, 6, 4, 0]` pues:
-  `8` está en el lugar `0` que es múltiplo de `3` y `7`, 
- `11` está en el lugar `3`, `21`en el lugar `6`, `31` en el lugar `9`, `6` en el lugar `12`y `0` en el lugar `15`,  todas ubicaciones múltiplos de `3`.
- Finalmente, en la lista también están  `22`,  en el lugar `7`, y `4` en el lugar `14`, que son ubicaciones múltiplos de  `7`. 



````python
# Escribir el código en esta celda
def extraer_coordenadas(lista):
  #pre: recibe una lista de numeros enteros
  #post: devuelve otra lista con los elementos de las coordenadas que son multiplos de 3 y 7
  assert type (lista) == list and all(type(x) == int for x in lista)

  lista_nueva = []

  for i in range(len(lista)):
    if i%3==0 or i%7==0:
      lista_nueva.append(lista[i])
  
  return lista_nueva
````

````python
# Tests: la función debe pasar los siguientes tests

test(1, extraer_coordenadas([8,5,3,11,0,1,21,22,23,31,5,6,6,3,4,0,3,2]), '[8, 11, 21, 22, 31, 6, 4, 0]')
test(2, extraer_coordenadas([50, 19, 25, 34, 10, 15, 20, 43, 39, 15, 36, 49, 3, 23, 30, 30, 20, 25, 5, 26, 26, 41, 15]), '[50, 34, 20, 43, 15, 3, 30, 30, 5, 41]')
test(3, extraer_coordenadas([7, 37, 1, 37, 19, 47, 41, 46, 5, 14, 7, 23, 29, 18, 7, 34, 46, 37, 39, 15, 21, 12, 38, 12, 20, 12, 24, 35, 16, 31]), '[7, 37, 41, 46, 14, 29, 7, 34, 39, 12, 20, 35, 16]')

````

**Ejercicio 4.** Definir una función `sin_repeticiones_consecutivas()` que dada una lista de números enteros devuelve una lista con los mismos enteros en el mismo orden salvo que evitando las repeticiones de números consecutivos: Por ejemplo

```
sin_repeticiones_consecutivas([8,8,5,5,5,8,3,11,0,0,6,6,3,3,6,6])
```

devuelve la lista `[8, 5, 8, 3, 11, 0, 6, 3, 6]`.


````python
# Escribir el código en esta celda
def sin_repeticiones_consecutivas(lista):
  #pre: recibe una lista de números enteros
  #post: devuelve una lista con los mismos enteros y en el mismo orden pero sin las repeticiones
  assert type(lista)==list and all(type(x)==int for x in lista), "Error, debe recibir una lista de números enteros"

  i = 1

  while i < len(lista):
    if lista[i-1] == lista[i]:
      lista.pop(i-1)
    else:
      i=i+1
  
  return lista
````

````python
# Tests: la función debe pasar los siguientes tests

test(1, sin_repeticiones_consecutivas([]), '[]')
test(2, sin_repeticiones_consecutivas([5]), '[5]')
test(3, sin_repeticiones_consecutivas([8,8,5,5,5,8,3,11,0,0,6,6,3,3,6,6]), '[8, 5, 8, 3, 11, 0, 6, 3, 6]')
````

**Ejercicio 5.** Implementar el tipo de datos ángulo, con su representación en grados sexagesimales en el rango `0°00'00"` a `359°59'59"`, con las operaciones: 
- `Angulo(a, b, c)`: constructor a partir de tres enteros no negativos `a`, `b` y `c`, que representa el ángulo en notación sexagesimal de `a` grados, `b` minutos y `c` segundos. También admite `Angulo(a, b)` que corresponde a `a` grados, `b` minutos y 0 segundos, y `Angulo(a)` que corresponde a `a` grados, 0 minutos y 0 segundos. Por ejemplo, `Angulo(780, 59, 90)` construye la representación de `61°00'30"`.
- `+`: suma dos ángulos, devolviendo otro ángulo con la suma de los dos ángulos dados. Por ejemplo, la suma de `90°00'00"` y `270°00'00"` es `0°00'00"`.
- `- angulo`: que devuelve otro ángulo con el opuesto aditivo de `angulo`. Por ejemplo, el opuesto de `90°00'00"` es `270°00'00"`.
- `-`: que resta dos ángulos, devolviendo otro ángulo con la resta de dos ángulos dados. Como es habitual, esta operación puede definirse combinando las dos anteriores.
- `*`: multiplica un ángulo por un entero no negativo. Por ejemplo, `89°59'59"` * `5` es `89°59'55"`.
- `==`:  determina si dos ángulos son en realidad el mismo ángulo,
- `<=`: determina si un ángulo es menor o igual que otro,
- `print`: imprime `g°mm'ss"`. 

Todos los métodos que construyen o devuelven ángulos, deben devolverlo en su representación en el rango `0°00'00"` a `359°59'59"`. Para ello, se sugiere definir un método `__simplify()` que lleva el ángulo a dicho rango e invocarla al final de cada método salvo el constructor.

*Ayuda.* 
- `__add__`:  es el método que permite definir `+`, la suma de dos elementos,
- `__neg__`:  es el método que permite definir el operador unario  `-`, el opuesto de un elemento,
- `__sub__`:  es el método que permite definir el operador binario `-`, la resta de dos elementos,
- `__mul__`:  es el método que permite definir  `*`, la multiplicación de dos elementos,
- `__eq__`:  es el método que permite definir  `==`.
- `__le__`:  es el método que permite definir  `<=`.


````python
class Angulo:

    def __init__(self, grados : int, minutos : int = 0, segundos : int = 0):
        # constructor de ángulo, "normaliza" para representarlo como un valor entre 0°00'00" y 359°59'59"
        assert grados >= 0 and minutos >= 0 and segundos >= 0,'Error: el ángulo debe ser no negativo'
        self.__segundos = segundos % 60
        self.__minutos = minutos + segundos // 60
        self.__grados = grados + self.__minutos // 60
        self.__minutos = self.__minutos % 60
        self.__grados = self.__grados % 360

     

    def __str__(self) -> str:
        # imprime el ángulo con la notación g°mm'ss"
        return str(self.__grados) + '°' + '{:0>2}'.format(self.__minutos) + "'" + '{:0>2}'.format(self.__segundos) + '"'

    def __simplify(self, grados: int, minutos:int, segundos:int):
        
        self.__segundos = segundos % 60
        self.__minutos = minutos + segundos // 60
        self.__grados = grados + self.__minutos // 60
        self.__minutos = self.__minutos % 60
        self.__grados = self.__grados % 360
        
        return (str(self.__grados) + "°" + str(self.__minutos) + "´" + str(self.__segundos) + '"')


    def __add__(self, otro):
        # Completar la definición, no debe modificar los operandos, debe devolver un nuevo ángulo
        assert type(otro) == Angulo, 'otro debe ser un ángulo'
      
        suma_grado = self.__grados + otro.__grados
        suma_min = self.__minutos + otro.__minutos
        suma_seg = self.__segundos + otro.__segundos

        return Angulo.__simplify(suma_grado, suma_min, suma_seg)

    def __neg__(self):
        # devuelve el opuesto de un ángulo restandolo de 360°
        a = 359 - self.__grados
        b = 59 - self.__minutos
        c = 60 - self.__segundos
        res = Angulo(a, b, c)
        return res

    def __sub__(self, otro):
        # Completar la definición, no debe modificar los operandos, debe devolver un nuevo ángulo
        assert type(otro) == Angulo, 'otro debe ser un ángulo'

        otro_neg = Angulo.__neg__(otro)
        resta = Angulo.__add__(self,otro_neg)
        return Angulo.__simplify(resta)
    
    def __mul__(self, k : int):
        # Completar la definición, no debe modificar los operandos, debe devolver un nuevo ángulo, k debe ser no negativo

        mul_grado = self.__grados * k
        mul_min = self.__minutos * k
        mul_seg = self.__segundos * k

        res = (mul_grado, mul_min, mul_seg)
        
        return Angulo.__simplify(res)

    def __eq__(self, otro) -> bool:
        # Completar la definición, no debe modificar los operandos
        assert type(otro) == Angulo, 'otro debe ser un ángulo'

        return self.__grados == otro.__grados and self.__minutos == otro.__minutos and self.__segundos == otro.__segundos

    def __le__(self, otro) -> bool:
        # Completar la definición, no debe modificar los operandos
        assert type(otro) == Angulo, 'otro debe ser un ángulo'

        return self.__grados <= otro.__grados or (self.__grados == otro.__grados and self.__minutos <= otro.__minutos) or (self.__grados == otro.__grados and self.__minutos == otro.__minutos and self.__segundos <= otro.__segundos)
````

````python
# Tests: los métodos deben pasar los siguientes tests

cero = Angulo(0)
uno = Angulo(1)
casirecto = Angulo(89,59,59)
recto = Angulo(90)
masquerecto = Angulo(90,0,1)
casillano = casirecto * 2
llano = Angulo(180)
masquellano = masquerecto + recto
chico = masquerecto - casirecto
grande = casirecto - masquerecto
completo = Angulo(360)

test(1, cero, '0°00\'00"')
test(2, uno, '1°00\'00"')
test(3, casirecto, '89°59\'59"')
test(4, recto, '90°00\'00"')
test(5, masquerecto, '90°00\'01"')
test(6, casillano, '179°59\'58"')
test(7, llano, '180°00\'00"')
test(8, masquellano, '180°00\'01"')
test(9, -casillano, '180°00\'02"')
test(10, chico, '0°00\'02"')
test(11, grande, '359°59\'58"')
test(12, completo, '0°00\'00"')
test(13, cero == completo, 'True')
test(14, llano <= llano, 'True')
test(15, casillano <= llano, 'True')
test(16, llano <= casillano, 'False')
````
